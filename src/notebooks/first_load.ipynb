```python
def get_start_sz_time(sz_info: pd.DataFrame, edf: str) -> float:
    """Get start time of seizure for given edf file.
    """
    import ast
    
    char_info = sz_info[sz_info['file_name'] == edf]['start_end_times'].to_numpy()[0]
    print(char_info)
    int_info = ast.literal_eval(char_info)[0][0]
    return int_info

sz_7_start_times = dict(zip(sz_info_7["file_name"].values, [0.0]*len(sz_info_7["file_name"].values)))

for edf in sz_info_7["file_name"].values:
    start_time = get_start_sz_time(sz_info_7, edf)
    print(start_time)
    sz_7_start_times[edf] = start_time
```
