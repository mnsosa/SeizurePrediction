```python
# How many edf we have?

patients = data_handler.get_patients()
print(f'Number of patients: {len(patients)}')

edf_files = {}

for patient in patients:
    edf_files[patient] = data_handler.get_patient_edf(patient)

print(f'Number of edf files: {sum([len(edf_files[patient]) for patient in edf_files])}')
```

```python
# how is the distribution of edf files per patient?

for patient in edf_files:
    print(f'Patient {patient} has {len(edf_files[patient])} edf files')
```

```python
# bar plot of edf files per patient

import matplotlib.pyplot as plt


plt.style.use('ggplot')

# font dict
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

# font dict for title
font_title = {
    'family': 'serif',
    'color':  'darkred',
    'weight': 'normal',
    'size': 20,
}


plt.figure(figsize=(15, 6))
plt.bar(edf_files.keys(), [len(edf_files[patient]) for patient in edf_files])
plt.title('Number of edf files per patient', fontdict=font_title)
plt.xlabel('Patient', fontdict=font)
plt.ylabel('Number of edf files', fontdict=font)
# font dict
# rotate 45 ticks
plt.xticks(rotation=45)
plt.tick_params(axis='both', which='major', labelsize=10, pad=10)
# more space between ticks and xlabel and ylabel
plt.show()
```
