# Using data_handler.py

The main idea is to read the edf files and extract the data from the summary files too. This is a simple demo.

```python
from sz_utils import data_handler

# get all patient labels
patients = data_handler.get_patients()
print("Patients\n", patients)

# get all edf files in a patient
edf_files = data_handler.get_patient_edf(patients[0])
print(f"\nEDF files in patient {patients[0]}\n", edf_files)

# get the summary of a patient
summary = data_handler.get_summary(patients[0])
print(f"\nSummary of patient {patients[0]} (first 200 chars)\n", summary[:200])
```

```python
# get the raw edf data
edf_data = data_handler.get_edf_data(patients[0], edf_files[0])
print(f"\nEDF Data of patient {patients[0]} and file {edf_files[0]} (first 5 rows)\n")
edf_data.head()
```

```python
# get the seizure data
seizure_data = data_handler.get_seizure_data(patients[0])
seizure_data.head()
```

```python
# filter seizure data by number of seizures > 0
seizure_data[seizure_data["number_of_seizures"] > 0]
```

```python
# the start and end times has a str format.
# an example on how to convert it to a list of tuples
example = seizure_data[seizure_data["number_of_seizures"] > 0]["start_end_times"]
example = example.values[0]

print(type(example))
example
```

```python
# convert the str '[(2996, 3036)]' to a list of tuples
import ast

result = ast.literal_eval(example)
print(type(result))
print(result)
print(result[0][0])
print(type(result[0][0]))
```
